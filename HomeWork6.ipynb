```python
import pandas as pd
```

```python
#Task 1

stats = pd.read_csv('keywords.csv')
stats.head()
```

```python
stats['word_count'] = stats['keyword'].apply(lambda x: len(x.split(' ')))
```

```python
def geo_class (row):
    for x in range(row['word_count']):
        if row['keyword'].split(' ')[x]=='москва' or row['keyword'].split(' ')[x]=='тула' or row['keyword'].split(' ')[x]=='ярославль':
            region = 'центр'
        elif row['keyword'].split(' ')[x]=='петербург' or row['keyword'].split(' ')[x]=='псков' or row['keyword'].split(' ')[x]=='мурманск':
            region = 'Северо-Запад'
        elif row['keyword'].split(' ')[x]=='владивосток' or row['keyword'].split(' ')[x]=='сахалин' or row['keyword'].split(' ')[x]=='хабаровск':
            region = 'Дальний Восток'    
        else:
            region = 'undefined'
    return region
```

```python
stats['region'] = stats.apply(geo_class, axis=1)
stats
```

```python
stats_filter = stats[stats['region']!='undefined']
stats_filter
```

```python
stats['region'].unique()
```

```python
# Task 2

stats = pd.read_csv('ratings.csv')
stats.head()
```

```python
def define_class(row):
    if row['rating'] <= 2.0:
        class_name = 'низкий рейтинг'
    elif row['rating'] <= 4.0 and row['rating'] > 2.0:
        class_name = 'средний рейтинг'
    else:
        class_name = 'высокий рейтинг'
    return class_name
```

```python
stats['class_categ'] = stats.apply(define_class, axis=1)
stats.head()
```

```python
stats['class_categ'].unique()
```

```python
stats_filter = stats[stats['class_categ']=='высокий рейтинг']
stats_filter.head()
```

```python
stats_filter['rating'].unique()
```

```python
# Task 3

ratings = pd.read_csv('ml-latest-small/ratings.csv')
ratings.head()
```

```python
ratings_count = ratings.groupby('userId').count().reset_index()[['userId', 'movieId']]
ratings_count.head()
```

```python
film_fans_ratings_count = ratings_count[ ratings_count['movieId'] >= 100 ]
film_fans_ratings_count.head()
```

```python
film_fans_user_ids = film_fans_ratings_count['userId'].tolist()
```

```python
fans_data = ratings[ ratings['userId'].isin(film_fans_user_ids) ]
fans_data.head()
```

```python
fans_data_group = fans_data.groupby('userId').agg([min,max])
```

```python
import time
from datetime import date
```

```python
from datetime import datetime
```

```python
def define_day(row):
    return (datetime.fromtimestamp(row['timestamp']['max']) - datetime.fromtimestamp(row['timestamp']['min'])).days
```

```python
fans_data_group['days'] = fans_data_group.apply(define_day, axis=1)
fans_data_group.head()
```
