```python
#Task 4

data = pd.read_csv('power.csv')
```

```python
filtered_countries_categories = data[ ((data['country']=='Lithuania') | (data['country']=='Estonia') | (data['country']=='Latvia')) & ((data['category']==4) | (data['category']==12) | (data['category']==21))]
```

```python
filtered_year_2005 = filtered_countries_categories[ filtered_countries_categories['year'] >= 2005]
```

```python
filtered_year_2010 = filtered_year_2005[ filtered_year_2005['year'] <= 2010]
```

```python
filtered_quantity = filtered_year_2010[ filtered_year_2010['quantity'] >= 0]
```

```python
sum(filtered_quantity['quantity'])
```
