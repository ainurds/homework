```python
categorial_cols = [
    'action_type', 'combined_shot_type', 'period', 'season', 'shot_type',
    'shot_zone_area', 'shot_zone_basic', 'shot_zone_range', 'game_year',
    'game_month', 'opponent', 'loc_x', 'loc_y']

for cc in categorial_cols:
    dummies = pd.get_dummies(data_cl[cc])
    dummies = dummies.add_prefix("{}#".format(cc))
    data_cl.drop(cc, axis=1, inplace=True)
    data_cl = data_cl.join(dummies)
```

```python
data_submit = data_cl[unknown_mask]

X = data_cl[~unknown_mask]
Y = target[~unknown_mask]
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import roc_auc_score, roc_curve
```

```python
linreg = LinearRegression(normalize=True)
linreg.fit(X, Y)
y_pred = linreg.predict(X)

roc_auc_pred = np.round(roc_auc_score(Y, y_pred), 2)
roc_auc_pred
```

```python
ret = [roc_auc_pred]
ret.extend([linreg.intercept_])
ret.extend(linreg.coef_)
```

```python
col = ['roc_auc_pred', 'intercept'] + [format(x) for x in X.columns]
ind = ['y_pred']
coef_matrix_simple = pd.DataFrame(index=ind, columns=col)

coef_matrix_simple.iloc[0, ] = ret
```

```python
from sklearn.linear_model import Ridge
```

```python
def ridge_regression(alpha):
    # Fit
    ridgereg = Ridge(alpha=alpha, normalize=True)
    ridgereg.fit(X, Y)
    y_pred_ridge = ridgereg.predict(X)
    
    # format
    roc_auc_ridge = np.round(roc_auc_score(Y, y_pred_ridge), 2)
    ret = [roc_auc_ridge]
    ret.extend([ridgereg.intercept_])
    ret.extend(ridgereg.coef_)
    return ret
```

```python
alpha_ridge = [1e-15, 1e-10, 1e-8, 1e-4, 1e-3, 1e-2, 1, 5, 10, 20]
```

```python
col = ['roc_auc_ridge', 'intercept'] + [format(x) for x in X.columns]
ind = ['alpha_%.2g' % alpha_ridge[i] for i in range(0, 10)]
coef_matrix_ridge = pd.DataFrame(index=ind, columns=col)

for i in range(10):
    coef_matrix_ridge.iloc[i, ] = ridge_regression(alpha_ridge[i])
```
